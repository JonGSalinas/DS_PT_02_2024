# FAANG Historical Stock Price Data

About Dataset

FAANG STOCKS ---> Stocks data of Facebook, Amazon, Apple, Netflix & Google companies.
CONTENT :

The dataset contains:
-- The data that was scraped from the yahoo finance website.
-- Open, Close, Adj Close, High, Low stock prices and Volume of Facebook, Amazon, Apple, Netflix & Google stocks for the last 10 years.
This dataset contains the stock data of Amazon for the last 5 years (from 2016 to date). The insights from this data can be used to perform data analysis, backtesting various trading strategies & stock price trend prediction. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from os import listdir
from os.path import isfile, join


```

```python
path = "./data"
files = [f for f in listdir(path) if isfile(join(path, f))]
files
```

```python
faang = pd.DataFrame()
for f in files:
    tmp = pd.read_csv(f"./data/{f}")
    # nueva columna Nombre con el nombre de la compa√±ia
    tmp["Nombre"] = f.split("_")[0]
    faang = pd.concat([faang,tmp])
```

```python
faang.to_csv("faang.csv")
```
